```python
import sqlite3
import pandas as pd
import FinanceDataReader as fdr
```

```python
df_krx = fdr.StockListing("KRX")

conn = sqlite3.connect('./complete_db.stock')
c = conn.cursor()
data = pd.read_sql('SELECT * FROM article', con=conn)
c.close()

data_copy = data[['title', 'stock']].copy()


for i in range(0,len(df_krx)):
    for j in range(0,len(data)):
        if df_krx["Name"][i] in data_copy["title"][j]:
            data_copy.iloc[j]['stock'] = df_krx["Name"][i]

data_copy
```

```python
non_stock = data.drop(['stock'], axis=1)

add_stock = pd.merge(non_stock, data_copy, on='title')
add_stock
```

```python
con = sqlite3.connect('./stock_db.stock')
c = con.cursor()
add_stock.to_sql('article', con, index=False)
c.close()
```

```python
conn = sqlite3.connect('./stock_db.stock')
c = conn.cursor()
st_data = pd.read_sql('SELECT * FROM article', con=conn)
c.close()

st_data.to_csv('./score_match.csv', index=False, header=True, encoding='utf-8-sig')
```
