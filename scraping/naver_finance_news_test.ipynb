```python
from selenium import webdriver
```

```python
from bs4 import BeautifulSoup
```

```python
driver = webdriver.Chrome('./chromedriver.exe')
```

```python
driver.get('https://finance.naver.com/news/market_special.nhn')
```

```python
elem = driver.find_element_by_xpath('//*[@id="contentarea_left"]/div[2]/table/tbody/tr[1]/td[1]/a')
elem.click()
```

```python
html = driver.page_source
html
```

```python
soup = BeautifulSoup(html, 'html.parser')
soup
```

```python
datas = soup.select('div.article_info > h3')
data = datas[0]
data
```

```python
data.text.strip()
```

```python
presses = soup.select('span.press > img')
press = presses[0]
press
```

```python
press['title']
```

```python
dates = soup.select('div.article_sponsor > span')
date = dates[0]
date
```

```python
date.text.strip()
```

```python
contents = soup.select('div#content')
content = contents[0]
content
```
