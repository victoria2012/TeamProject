```python
import sqlite3
import pandas as pd
```

```python
with open("./negative_words.txt", encoding='utf-8') as neg:
    negative = neg.readlines()
    
with open("./positive_words.txt", encoding='utf-8') as pos:
    positive = pos.readlines()
    
negative = [neg.replace("\n", "") for neg in negative]
positive = [pos.replace("\n", "") for pos in positive]
```

```python
conn = sqlite3.connect('./stock_db.stock')
c = conn.cursor()
data = pd.read_sql('SELECT * FROM article', con=conn)
c.close()
```

```python
label = list()
title_copy = data[['title']].copy()

for i in range(0,len(title_copy)):
    cnt = 0
    for j in range(0,len(positive)):
        if positive[j] in title_copy['title'][i]:
            cnt += 1
    for k in range(0,len(negative)):
        if negative[k] in title_copy['title'][i]:
            cnt -= 1
    
    if cnt >= 1:
        cnt = 2
        label.append(cnt)
    elif cnt == 0:
        cnt = 1
        label.append(cnt)
    else:
        cnt = 0
        label.append(cnt)
    
title_copy['label'] = label
title_copy
```

```python
add_label = pd.merge(data, title_copy, on='title')
add_label
```

```python
con = sqlite3.connect('./label_db.sqlite3')
c = con.cursor()
add_label.to_sql('article', con, index=False)
c.close()
```
